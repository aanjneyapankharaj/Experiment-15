```python
# Tokenization
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize

text="Natural language processing (NLP) is a field of computer science, artificial intelligence."
tokens=word_tokenize(text)
print(tokens)
```

```python
# Sentence Tokenization
from nltk.tokenize import sent_tokenize
text="Hello everyone . You are studying NLP article"
sentences=sent_tokenize(text)
print(sentences)
```

```python
# Stop word removal
# Stop word such as is,the,and,in are removed.
from nltk.corpus import stopwords
stop_words=set(stopwords.words('english'))
words=word_tokenize(text)
filtered_words=[word for word in words if word.lower() not in stop_words]
print(filtered_words)
```

```python
# Stemming
# Stemming reduces words to their root form.
# Stemming is done from the end.
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
words=["Football","F1","Basketball","Cricket","Playing","Singing"]
for w in words:
  print(w,":",stemmer.stem(w))
```

```python
# Lemmanization
# Lemmanization converts words to their dictionary form.
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
print(lemmatizer.lemmatize("Better"))
print(lemmatizer.lemmatize("Eating"))
```

```python
# Part of speech (POS) Tagging
# POS tagging identifies grammatical roles of words
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk import pos_tag
words=word_tokenize(text)
pos_tags=pos_tag(words)
print(pos_tags)
```
